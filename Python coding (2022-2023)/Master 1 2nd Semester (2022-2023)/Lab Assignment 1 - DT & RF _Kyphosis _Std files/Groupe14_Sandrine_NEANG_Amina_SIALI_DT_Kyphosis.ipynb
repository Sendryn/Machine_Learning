We are going to train a model that can predict wether a child has Kyphosis after their spinal surgery. We will be using a single decision tree and random forests. 

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Needed to hide warnings in the matplotlib sections
import warnings
warnings.filterwarnings("ignore")
```

## Get the Data
Now we import the data from the CSV file into a dataframe. The attributes of our set are the Age of the child (in months), the Number (of vertebrae) and the Start (top vertebra) operated on.

```python
df = pd.read_csv('kyphosis.csv')
```

```python
df.head() 
#head() shows the first rows and columns of the dataframe
```

```python
df.info() 
#info() shows the summary of the Kyphosis dataset.
#The summary includes list of all columns with their data types and the number of non-null values in each column. 
#We also have the value of range index provided for the index axis.
```

```python
df["Kyphosis"].value_counts()
#value_counts() counts the number of "absent" and "present" in Kyphosis variable.
```

## Exploratory Data Analysis (EDA)

We'll just check out a simple pairplot for this small dataset.

```python
sns.pairplot(df, hue = "Kyphosis", palette = "Set1")
#It allows us to look at how Kyphosis variable is distributed and related to others variables
```

## Train Test Split

Let's split up the data into a training set and a test set.

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
X_train.head()
```

## Decision Trees

We'll start just by training a single decision tree.

Now we choose from sklearn.tree the decision tree classifier model

```python
from sklearn.tree import DecisionTreeClassifier
```

Create a model called dtree 

```python
dtree = DecisionTreeClassifier()
```

Fit the model to the training data

```python
dtree.fit(X_train,y_train)
#fit() is used for training the model
#we only use training data because we don't want to bias our model with information from the test data.
```

## Prediction and Evaluation 

Let us evaluate now, how well our decision tree model is able to predict based on the attributes (Age, Number, Start), whether or not Kyphosis is present at the child or not.

```python
predictions1 = dtree.predict(X_test)
#the prediction is done with predict() based on the attributes from the test data to see if the child has Kyphosis or not
```

sklearn.metrics provides the function confusion_matrix that compares the predicted values with the actual lable values, and outputs the confusion_matrix.

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions1))
#it shows the confusion matrix of the prediction1
#12 is the number of true positive (children who have the Kyphosis)
#2 is the number of true negative (children who don't have the Kyphosis)
#5 and 6 are the number of false positive and false negative.
```

Similarly, sklearn.metrics provides the function classification_report that outputs the main classification metrics: such as precision, recall, f1-score, accuracy, etc. You can read more about it in the sklearn documantation here https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html

```python
print(classification_report(y_test,predictions1))
#The model predict absent Kyphosis with a precision percentage of 0.67 
#which is better than the precision percentage of the present Kyphosis.

#We can see that the f1-score of absent Kyphosis is higher and closer to one than the present Kyphosis, 
#then the model have better prediction for absent Kyphosis than present Kyphosis.
```

## Random Forests

#### Now let's compare the decision tree model to a random forest.
A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting. The sub-sample size is always the same as the original input sample size but the samples are drawn with replacement. You can read more about it here https://scikit-learn.org/stable/modules/ensemble.html#forest

From sklearn.ensemble, choose the Random Forest classifier. Note that, this classifier is under ensemble, since it builds an emsemble of trees, so it falls under the category of ensemble learning. 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
predictions2 = rfc.predict(X_test)
#fit() is used for training the model
#we only use training data because we don't want to bias our model with information from the test data.
```

```python
print(confusion_matrix(y_test,predictions2))
#it shows the confusion matrix of the prediction2
#17 is the number of true positive (children who have the Kyphosis)
#2 is the number of true negative (children who don't have the Kyphosis)
#0 and 6 are the number of false positive and false negative.
```

```python
print(classification_report(y_test,predictions2))
#The model predict present Kyphosis with a precision percentage of 1.00 
#which is better than the precision percentage of the absent Kyphosis of 0.74.

#We can see that the f1-score of absent Kyphosis is higher and closer to one than the present Kyphosis, 
#then the model have better prediction for absent Kyphosis than present Kyphosis.
```
